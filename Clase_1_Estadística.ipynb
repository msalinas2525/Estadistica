```python
#Punto1

import matplotlib.pyplot as plt
import seaborn as sns
# Variables categóricas
categorical_cols = ["GENERO", "OCUPACION", "PLAN_BENEFICIOS"]

for col in categorical_cols:
    plt.figure(figsize=(10, 6))

    # Diagrama de barras
    sns.countplot(data=df, y=col, order=df[col].value_counts().index)
    plt.title(f'Distribución de {col} (Barras)')
    plt.xlabel('Frecuencia')
    plt.ylabel(col)
    plt.show()

    # Diagrama de torta (for columns with reasonable number of unique values)
    if df[col].nunique() < 20: # Avoids too many slices in pie chart
        plt.figure(figsize=(8, 8))
        df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
        plt.title(f'Distribución de {col} (Torta)')
        plt.ylabel('') # Remove default y-label
        plt.show()
    else:
        print(f"Skipping pie chart for {col} due to too many unique values.")
```

```python
# Punto2
numerical_cols = ["EDAD", "PESO", "TALLA"]

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribución de {col} (Histograma)')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, y=col)
    plt.title(f'Distribución de {col} (Diagrama de Caja)')
    plt.ylabel(col)
    plt.show()
```

```python
# Punto 3: Eliminar outliers usando IQR
numerical_cols = ["EDAD", "PESO", "TALLA"]
df_no_outliers = df.copy()

for col in numerical_cols:
    Q1 = df_no_outliers[col].quantile(0.25)
    Q3 = df_no_outliers[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_no_outliers = df_no_outliers[(df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)]

print("Información del dataset después de eliminar outliers:")
print(df_no_outliers.info())
print("Primeras filas del dataset sin outliers:")
print(df_no_outliers.head())
```

```python
# Punto 4: Generar histogramas para datos sin outliers
numerical_cols = ["EDAD", "PESO", "TALLA"]

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df_no_outliers, x=col, kde=True)
    plt.title(f'Distribución de {col} sin Outliers (Histograma)')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
# Punto 5: Análisis de normalidad
import scipy.stats as stats
import matplotlib.pyplot as plt

numerical_cols = ["EDAD", "PESO", "TALLA"]

for col in numerical_cols:
    print(f"\nAnálisis de normalidad para la columna: {col}")

    # Shapiro-Wilk test
    shapiro_test = stats.shapiro(df_no_outliers[col])
    print(f"Shapiro-Wilk test: Statistics={shapiro_test.statistic:.3f}, p-value={shapiro_test.pvalue:.3f}")

    # Kolmogorov-Smirnov test (using Lilliefors for estimated parameters)
    kolmogorov_test = stats.kstest(df_no_outliers[col], cdf='norm')
    print(f"Kolmogorov-Smirnov test: Statistics={kolmogorov_test.statistic:.3f}, p-value={kolmogorov_test.pvalue:.3f}")

    # Anderson-Darling test
    anderson_test = stats.anderson(df_no_outliers[col], dist='norm')
    print(f"Anderson-Darling test: Statistics={anderson_test.statistic:.3f}")
    for i in range(len(anderson_test.critical_values)):
        sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
        print(f'    {sl}%: {cv:.3f} {"(Data is normal at this level)" if anderson_test.statistic < cv else "(Data is not normal at this level)"}')

    # Jarque-Bera test
    # This test is not available in scipy.stats for direct use on a single variable's distribution,
    # but it's often used in regression analysis context.
    # For a single variable, checking skewness and kurtosis is related.
    print("Jarque-Bera test is typically used in regression contexts. Analyzing skewness and kurtosis instead:")
    print(f"Skewness: {df_no_outliers[col].skew():.3f}")
    print(f"Kurtosis: {df_no_outliers[col].kurtosis():.3f}")


    # QQ plot
    plt.figure(figsize=(8, 6))
    stats.probplot(df_no_outliers[col], dist="norm", plot=plt)
    plt.title(f'QQ Plot for {col}')
    plt.show()
```
